```python
#importing libraries

import numpy as np
import matplotlib.pyplot as plt
import cv2


%matplotlib inline
```

```python
#importing the image

image=cv2.imread("/home/anurag/Downloads/duck.jpeg")

#Creating a copy of image

copy_image=np.copy(image)

plt.imshow(copy_image)
```

```python
 #convertng image from BGR color-space to RGB color-space

rgb_copy=cv2.cvtColor(copy_image,cv2.cv2.COLOR_BGR2RGB)

plt.imshow(rgb_copy)
```

```python
#Converting image to Gray scale for edge detection

gray_copy=cv2.cvtColor(rgb_copy,cv2.COLOR_RGB2GRAY)

plt.imshow(gray_copy,cmap="gray")
```

```python
#creating custom weight/kernal for horizontal and vertcal edge detection

h_kernal=np.array(  [[-1,0,1],
                      [-2,0,2],
                      [-1,0,1]])

v_kernal=np.array( [[-1,-2,-1],
                    [0,0,0],
                    [1,2,1]] )
```

```python
 # Using Convolutional 2D-Filter 

h_filter_copy=cv2.filter2D(gray_copy,-1,h_kernal)

v_filter_copy=cv2.filter2D(gray_copy,-1,v_kernal)
```

```python
#ploting the horizontal filtered image

plt.imshow(h_Filter_copy,cmap="gray")
```

```python
#ploting the vertical filtered image

plt.imshow(v_filter_copy,cmap="gray")
```
